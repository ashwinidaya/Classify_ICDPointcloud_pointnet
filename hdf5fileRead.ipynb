```
#ICD
import h5py
import numpy as np
from plyfile import PlyData, PlyElement

filenames = [line.rstrip() for line in open("/content/drive/My Drive/Colab Notebooks/ICD_PointCloud/data/filelist", 'r')]




```

```
#f = h5py.File("./hdf5_data/data_training.h5", 'w')
f = h5py.File("/content/drive/My Drive/Colab Notebooks/ICD_PointCloud/data/hdf5_data/dataTry1.h5", 'w')

a_data = np.zeros((len(filenames), 2048, 3))
label = np.zeros((len(filenames), 2048), dtype = np.uint8)
```

```
#label=np.random(2048)
for i in range(0, len(filenames)):
	plydata = PlyData.read("/content/drive/My Drive/Colab Notebooks/ICD_PointCloud/data/ply/" + filenames[i] + ".ply")
	piddata = [line.rstrip() for line in open("/content/drive/My Drive/Colab Notebooks/ICD_PointCloud/data/points_label/" + filenames[i] + ".seg", 'r')]
	for j in range(0, 2048):
		a_data[i, j] = [plydata['vertex']['x'][j], plydata['vertex']['y'][j], plydata['vertex']['z'][j]]
		label[i,j] = piddata[j]


data = f.create_dataset("data", data = a_data)
label = f.create_dataset("label", data = sum(label))
```
