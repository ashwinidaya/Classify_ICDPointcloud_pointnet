```
import numpy as np
from pyntcloud import PyntCloud

cloud = PyntCloud.from_file("/content/b1.ply")

# Should show 4 red points, 1 blue, 1 green.
cloud.plot()

```

```
point_to_change = 0
new_rgb = np.array([255, 255, 255], dtype=np.uint8)

cloud.points.loc[point_to_change, ["red", "green", "blue"]] = new_rgb

# Should show 3 red points, 1 blue, 1 green and 1 white.
cloud.plot()
```

```
from pyntcloud import PyntCloud

cloud = PyntCloud.from_file("/content/bunny.ply")

cloud.add_scalar_field("hsv")

voxelgrid_id = cloud.add_structure("voxelgrid", n_x=32, n_y=32, n_z=32)

new_cloud = cloud.get_sample("voxelgrid_nearest", voxelgrid_id=voxelgrid_id, as_PyntCloud=True)

new_cloud.to_file("out_file.npz")
```

```
import open3d as o3d
from pyntcloud import PyntCloud

# FROM Open3D
original_triangle_mesh = o3d.io.read_triangle_mesh("/content/bunny.ply")
cloud = PyntCloud.from_instance("open3d", original_triangle_mesh)

# TO Open3D
cloud = PyntCloud.from_file("/content/bunny.ply")
converted_triangle_mesh = cloud.to_instance("open3d", mesh=True)  # mesh=True by default
```
